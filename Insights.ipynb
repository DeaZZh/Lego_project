```python
result.drop(['Unnamed: 0'],axis=1)
```

```python
result['Date Created']=pd.to_datetime(result['Date Created'])
result['Date Created']=result['Date Created'].dt.date
```

```python
result.head()
```

```python
import matplotlib.pyplot as plt
```

```python
plt.figure(figsize=(10,10))
result[result['sentiment']=='Negative'].groupby(['Date Created']).sentiment.count().plot(color='red',xlabel="Time Period")
result[result['sentiment']=='Positive'].groupby(['Date Created']).sentiment.count().plot(color='green',xlabel="Time Period")
result[result['sentiment']=='Neutral'].groupby(['Date Created']).sentiment.count().plot(color='blue',xlabel="Time Period")
plt.legend(["Negative", "Positive","Neutral"]);
```

```python
df = result[pd.to_datetime(result['Date Created']).dt.month == 4]
```

```python
filtered_df = df.loc[(df['Date Created'] >= pd.to_datetime('2022-04-01'))
                     & (df['Date Created'] < pd.to_datetime('2022-04-8'))]
```

```python
from wordcloud import WordCloud, STOPWORDS
import matplotlib.pyplot as plt

customStopwords=list(STOPWORDS)+ ['lego','twitter','de','euch','kelar']
wordcloudimage = WordCloud(
                          max_words=100,
                          max_font_size=500,
                          font_step=2,
                          stopwords=customStopwords,
                          background_color='white',
                          width=1000,
                          height=720
                          ).generate(Text)
plt.figure(figsize=(15,7))
plt.axis("off")
plt.imshow(wordcloudimage)
wordcloudimage
plt.show()
```
